#### Question_1:
__Alphabetic patterns__<br><br>
__Description__<br><br>
Given a positive integer 'n' less than or equal to 26, you are required to print the below pattern<br><br>
__Sample Input:__<br>
5 <br><br>
__Sample Output:__ <br>
```
--------e-------- 
------e-d-e------ 
----e-d-c-d-e---- 
--e-d-c-b-c-d-e-- 
e-d-c-b-a-b-c-d-e 
--e-d-c-b-c-d-e-- 
----e-d-c-d-e----

--------e-------- 
```

__Sample Input:__<br>
3 <br><br>
__Sample Output:__ <br>
```
----c---- 
--c-b-c-- 
c-b-a-b-c 
--c-b-c-- 
----c----
```

````python
# Solution_1: using slicing
n = int(input())

alpha = 'abcdefghijklmnopqrstuvwxyz'
alpha_list = [alpha[i] for i in range(n)]
num_list = list(range(n))
num_list = num_list[:-1]+num_list[::-1]

for i in num_list:
    alpha_seq = alpha_list[-(i+1):]
    row = alpha_seq[::-1]+alpha_seq[1:]
    print("-".join(row).center(n*4-3,"-"))
````

#### Question_5:
__Generalised Chocolate__<br><br>
__Description__<br><br>
Earlier you solved the chocolate problem where Sanjay had m rupees and cost of each chocolate was c rupees. Shopkeeper gave away one chocolate for three wrappers. In this problem lets generalise the question saying, Sanjay has m rupees, each chocolate costs c rupees, shopkeeper will give away k chocolates for w wrappers. Can you find now how many chocolates Sanjay will be able to eat?<br><br>
__Input:__<br>
 integers separated by commas in order m c w k<br>
integers c and w will be >0<br>
integers m and k will be >=0<br>
integer k will be <w <br><br>
__Output:__<br>
An integer denoting number of chocolates Sanjay will be able to get.<br><br>
__Sample Input:__<br>
15, 2, 3, 1<br><br>
__Sample Output:__ <br>
10<br><br>
__Sample Input:__<br>
15, 2, 3, 2<br><br>
__Sample Output:__ <br>
17<br><br>

````python
# Solution:
List1 = input().split(',')
m = int(List1[0]) # money
c = int(List1[1]) # chocolates
k = int(List1[2]) # wrappers to exchange
w = int(List1[3]) # how many chocolates one gets for exchanging 'k' wrappers

choc = m//c       # how many chocolate one can buy initially out of total money 'm', given the price of chocolate 'c'
wrapper = m//c    # same will be the number of wrappers initially

while wrapper//k != 0:
    wrap = (wrapper//k)*w
    choc += wrap
    wrapper = wrap + wrapper%k
print(choc)
````

#### Question_1:
__Lego Stack__<br><br>
__Description__<br><br>
You are given a row of Lego Blocks consisting of n blocks. All the blocks given have a square base whose side length is known. You need to stack the blocks over each other and create a vertical tower. Block-1 can go over Block-2 only if sideLength(Block-2)=>sideLength(Block-1).

From the row of Lego blocks, you can only pick up either the leftmost or rightmost block.

Print "Possible" if it is possible to stack all n cubes this way or else print "Impossible".<br><br>
__Input:__<br>
The input will contain a list of n integers representing the side length of each block's base in the row starting from the leftmost.<br><br>
__Output:__<br>
A String 'Possible' or 'Impossible'<br><br>
__Sample Input:__<br>
[5 ,4, 2, 1, 4 ,5]<br><br>
__Sample Output:__ <br>
Possible<br><br>

````python
# Solution_1: using while loop
import ast
side_list = ast.literal_eval(input())
n = len(side_list)
my_stack = []
i = 0
while i < n:
    if side_list[i] >= side_list[n-1]:
        my_stack.append(side_list[i])
        i += 1
    else:
        my_stack.append(side_list[n-1])
        n -= 1

i = 1
flag = False
while i < len(my_stack):
    if my_stack[i] > my_stack[i-1]:
        flag = True
    i += 1

if not flag:
    print("Possible")
else:
    print("Impossible")
````

````python
# Solution_2:
import ast
side_list = ast.literal_eval(input())
tower = []

while side_list:
    if side_list[0] >= side_list[-1]:
        tower.append(side_list[0])
        side_list.pop(0)
    else:
        tower.append(side_list[-1])
        side_list.pop(-1)
flag = False
for i in range(1,len(tower)):
    if tower[i] > tower[i-1]:
        print("Impossible")
        flag = True
        break
if not flag:        
    print("Possible")
    
````

A pascal's triangle is a very interesting mathematical concept.
Each number here is a sum of the two numbers directly above it.
Following is an 8 level Pascal's triangle:<br><br>

```
                        1 
                     1    1 
                  1    2    1 
               1    3    3    1 
            1    4    6    4    1 
         1    5    10    10    5    1 
      1    6    15    20    15    6    1 
```
You can read about Pascal's triangle [here](https://www.mathsisfun.com/pascals-triangle.html).<br>

The input will contain an integer n.<br>
The output will contain 1 line of the list of numbers representing the nth row of Pascal's triangle.<br><br>
__Sample Input:__<br>
6<br><br>
__Sample Output:__ <br>
[1, 5, 10, 10, 5, 1]<br><br>

````python
# Code for Pascal's Triangle
n = int(input())
for i in range(1,n+1):
    for j in range(n-i+1):
        print("  ",end=" ")
    c = 1
    for j in range(1,i+1):
        print("  ",c,sep=" ",end=" ")
        c = c*(i-j)//j
    print()
````

#### Question_4:
__Secret Message__<br><br>
__Description__<br><br>
You write all your passwords in a diary so that you don't forget them. But clearly this is too risky, so you came up with a simple plan, you will simply write it by shifting all the alphabets by a certain step. For eg: if you decide your step to be 3, then 'a' will become 'd', and 'k' will become 'n' and so for all alphabets. The last alphabets will simply circle back to 'a'. In this case, 'y' will become 'b' and so on. Now you just have to remember the step size, can then you can check the password anytime you want. You decided to write code to do this, now that you have learned coding in python. Your code will take in the step size and what is written in the diary and give out the real password.<br><br>

__Input:__<br>
A list of two elements.br>
The first element will be a string consisting of only alphabets which is taken from the diary and the second element will be the step size.<br><br>
__Output:__<br>
A string denoting the password<br><br>
__Sample Input:__<br>
['ytKveh', 4]<br><br>
__Sample Output:__ <br>
upGrad<br>
Explanation:<br>
The password was 'upGrad'. Upon changing it by step of 4,<br>
u became y,<br>
p became t,<br>
G became K,<br>
r became v,<br>
a became e,<br>
d became h,<br>
and thus what was written in the diary was ytKveh <br><br>
__Sample Input:__<br>
['banana', 7]<br><br>
__Sample Output:__ <br>
utgtgt<br><br>

````python
# Solution_1:
import ast
in_list = ast.literal_eval(input())
str1 = in_list[0]
n = int(in_list[1])
small_alpha = [chr(i) for i in range(ord('a'),ord('z')+1)]
cap_alpha = [chr(i) for i in range(ord("A"),ord("Z")+1)]
final_str = ''

for i in str1:
    if i in small_alpha:
        new_alpha = small_alpha[int(small_alpha.index(i) - n) % 26]
        final_str += new_alpha
    elif i in cap_alpha:
        new_alpha = cap_alpha[int(cap_alpha.index(i) - n) % 26]
        final_str += new_alpha
    else:
        final_str += i

print(final_str)
````

````python
# Solution_2:
# taking input using ast.literal_eval
import ast
in_list = ast.literal_eval(input())
in_str = in_list[0]
n = in_list[1]

# creating list for both upper and lower alphabets using list comprehension
small_alpha = [chr(ord('a') + i) for i in range(26)]
cap_alpha = [chr(ord('A') + i) for i in range(26)]

# creating empty string for concat
final_str = ""


for i in in_str:
    if i in small_alpha:
        new_alpha = chr(((ord(i) - ord("a") - n) % 26) + ord('a'))
        final_str += new_alpha
    elif i in cap_alpha:
        new_alpha = chr(((ord(i) - ord("A") - n) % 26) + ord('A'))
        final_str += new_alpha
    else:
        final_str += i
    
print(final_str)
````

#### Question_2:
__Maximum Distance__<br><br>
__Description__<br><br>
You will be given a list of repeated elements. You have to find the maximum distance between two same elements. The answer will be zero if there are no repeated elements.<br><br>
__Input:__<br>
A non-empty list of integers.<br><br>
__Output:__<br>
A single integer denoting the maximum distance between two same integers.<br><br>
__Sample Input:__<br>
[1, 2, 3, 2, 5, 1, 2, 4, 6, 2, 7, 8, 6]<br><br>
__Sample Output:__<br>
8<br><br>
Explanation:<br>
Max distance for 1: 5<br>
Max distance for 2: 8<br>
Max distance for 3: 0<br>
Max distance for 4: 0<br>
Max distance for 5: 0<br>
Max distance for 6: 4<br>
Max distance for 7: 0<br>
Max distance for 8: 0<br>

````python
# Solution_1:

# taking input by ast
import ast
in_list = ast.literal_eval(input())

# Creating an empty dictionary for distance
distance = {}

# The key of the dict is unique element in the in_list
# The value will be list of indexes which the key is present
# for that we need to iterate over in_list
# and appending the indexs of the elements as a list for values

for i in range(len(in_list)):
    if in_list[i] in distance:
        distance[in_list[i]].append(i) # here appending the list of indexes for that key
    else:
        distance[in_list[i]] = [i]

# Now the let be the in_list = [3, 2, 1, 2, 1, 4, 5, 8, 6, 7, 4, 2]
# The distance dictionary = {3: [0], 2: [1, 3, 11], 1: [2, 4], 4: [5, 10], 5: [6], 8: [7], 6: [8], 7: [9]}

# Now for getting the maximum distance need to take difference for -1th element to 0th element in values
max_distance = 0
for i in distance.values():
    diff = i[-1] - i[0]
    if diff > max_distance:
        max_distance = diff

print(max_distance)
````

#### Question_3:
__Vote for Food__<br><br>
__Description__<br><br>
Your team is going for camping and you are taking a vote to decide what food to pack for dinner.
Everyone gets a vote and the food item that gets at least one more than half of the votes wins. None of the items wins if nothing gets at least one more than half votes. Assume that every person gets only one vote.
The input will contain a list of food items where each occurrence of an item represents one vote. You should print the winning food item as output. If there is no clear winner, print "NOTA". <br><br>

__Sample Input:__<br>
["pasta","pasta","pasta","pasta","pasta","paratha","paratha","paratha"]<br><br>
__Sample Output:__<br>
pasta<br><br>
This question was asked in a Data Analyst interview.

````python
# Solution_1:
# Taking input by ast
import ast
food_list = ast.literal_eval(input())

# Creating an empty dictionary
# Key is the unique food
# value is the occurrence of food which is nothing but vote for the food
vote_dict = {}

# for that iterating over food_list key is the food and value will be occurrennce of the fooe
for food in food_list:
    if food not in vote_dict:
        vote_dict[food] = 1
    else:
        vote_dict[food] += 1

# for example the food_list = ["pasta","pasta","pasta","pasta","pasta","paratha","paratha","paratha"]
# the vote_dict = {'pasta': 5, 'paratha': 3}

# Now we need to see whether the max occurrence of the food is half of the vote,
# means we have to print "Nota"
# else we need to print max occurence key (here looping needed)

flag = False

for i in vote_dict.keys():
    if vote_dict[i] > len(food_list)/2:
        print(i)
        flag = True

if not flag:
    print("NOTA")
        
````

#### Question_4:
__Flatten a dictionary__<br><br>
__Description__<br><br>
Consider a nested dictionary as follows:

```
{'Fruit': 1, 'Vegetable': {'Cabbage': 2, 'Cauliflower': 3}, 'Spices': 4}
```

Your task is to flatten a nested dictionary and join the nested keys with the "_" character. For the above dictionary, the flattened dictionary would be as follows:

```
{'Fruit': 1, 'Vegetable_Cabbage': 2, 'Vegetable_Cauliflower': 3, 'Spices': 4}
```
__Input:__<br>
a nested dictionary<br><br>
__Output:__<br>
two lists. The first list will have keys and the second list should have values. Both lists should be sorted.<br><br>
__Sample Input:__<br>
```
{'Fruit': 1, 'Vegetable': {'Cabbage': 2, 'Cauliflower': 3}, 'Spices': 4}
```
__Sample Output:__<br>
```
['Fruit', 'Spices', 'Vegetable_Cabbage', 'Vegetable_Cauliflower']
[1, 2, 3, 4]
```

You will find the inbuilt function isinstance() of python quite useful here. Read more about it [here](https://python-reference.readthedocs.io/en/latest/docs/functions/isinstance.html).


````python
# Solution_1:

# Taking input by ast
import ast 
in_dict = ast.literal_eval(input())

# creating a user defined function of flatten_dict
def flatten_dict(d): # Here d is a dictionary
    item = {} # Creating an empty dictionary
    for key in d.keys():
        if isinstance(d[key],dict): # isinstance will check the value of d is dictionary
            for k in d[key].keys(): # checking the value dict
                item[key+"_"+k] = d[key][k] # replacing the value dict to key of the item dict by concat
        else:
            item[key] = d[key] # if value os the d is not dict means keep as it is
    return item

out1 = sorted(list(flatten_dict(in_dict).keys()))
out2 = sorted(list(flatten_dict(in_dict).values()))

print(out1)
print(out2)
````

````python
# Solution_2:

# Taking input by ast
import ast 
in_dict = ast.literal_eval(input())

# creating a user defined function of flatten_dict
def flatten_dict(d): # Here d is a dictionary
    def expand(key, value): # expanding dict as key,vlaue
        if isinstance(value,dict): # if value is dict
            return [(key+"_"+k,v) for k,v in flatten_dict(value).items()] # calling flatten_dict function by value dict
        else:
            return [(key,value)] # Keep as it is
    item = {k:v for k,v in d.items() for k,v in expand(k,v)} # creating new flatten_dict
    return item

# calling the flatten_dict keys and values
out1 = sorted(list(flatten_dict(in_dict).keys()))
out2 = sorted(list(flatten_dict(in_dict).values()))

print(out1)
print(out2)
````
